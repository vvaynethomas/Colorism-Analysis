```python
%matplotlib inline
import cv2
import os
import im_proc_functions
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
stats_df = im_proc_functions.batch_process('Images')
```

```python
stats_df
stats_df.sort_values(by='mean')
print(len(stats_df.index))
stats_df = stats_df.dropna()
print(len(stats_df.index))
```

```python
stats_df.set_index('image_name')
stats_df.describe()
```

```python
for col in ['mean','median','mode']:
  sns.distplot(stats_df[[col]])

stats_df[['mean','median','mode']].plot(kind='kde',subplots=True)
```

```python
r1_path = 'nicolekidman.jpg'
ref_image1 = cv2.imread(r1_path)
r1_1 = cv2.cvtColor(ref_image1, cv2.COLOR_BGR2RGB)
pixels1 = np.array(r1_1)
# plt.imshow(pixels1)
r1_face = im_proc_functions.extract_face(r1_path)
r1_2 = cv2.cvtColor(r1_face, cv2.COLOR_BGR2RGB)
pixels2 = np.array(r1_2)
# plt.imshow(pixels2)
r1_face_grey = cv2.cvtColor(r1_face, cv2.COLOR_BGR2GRAY)
r1_3 = cv2.cvtColor(r1_face_grey, cv2.COLOR_BGR2RGB)
pixels3 = np.array(r1_3)
# plt.imshow(pixels3)
fig, ax = plt.subplots(1,3)
ax[0].imshow(pixels1);
ax[1].imshow(pixels2);
ax[2].imshow(pixels3)
stats = im_proc_functions.get_greyscale_stats(r1_face)
print(stats)
```
