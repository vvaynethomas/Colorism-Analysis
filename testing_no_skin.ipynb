```python
%matplotlib inline
import im_proc_functions
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np
from random import randint
```

```python
test_face = np.nan
while not isinstance(test_face, np.ndarray):
    im_list = im_proc_functions.list_images('Images')
    test_image_path = im_list[randint(0,len(im_list))]
    print(test_image_path)
    test_image = cv2.imread(test_image_path)
    print(test_image.shape)
    image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
    pixels = np.array(image)
    plt.imshow(pixels)
    test_face = im_proc_functions.extract_face(test_image_path)
```
